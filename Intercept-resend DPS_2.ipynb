```python
#Alice encodes using phase encoding
def phase_encoding(bit1, bit2, circuit, qreg_q):
    encoding = []
    if bit1==0 and bit2==1: #message is 10, i.e,qiskit reads 01
        circuit.p(0, qreg_q[0])
        circuit.p(pi, qreg_q[1])
        circuit.p(pi, qreg_q[2])
        encoding.append('pi 0')
    elif bit1==1 and bit2==1:  #message is 11, i.e, phase difference is pi and pi
        circuit.p(pi, qreg_q[0])
        circuit.p(0, qreg_q[1])
        circuit.p(pi, qreg_q[2])
        encoding.append('pi pi')
    elif bit1==1 and bit2==0:  #message is 01, i.e, phase difference is 0 and 1
        circuit.p(0, qreg_q[0])
        circuit.p(0, qreg_q[1])
        circuit.p(pi, qreg_q[2])
        encoding.append('0 pi')
    elif bit1==0 and bit2==0:
        circuit.p(0, qreg_q[0])
        circuit.p(0, qreg_q[1])
        circuit.p(0, qreg_q[2])
        encoding.append('0 0')
    else:
        raise ValueError("Invalid input bits. Please provide valid values (0 or 1) for bit1 and bit2.")
    return circuit, encoding
```

```python
# def sample_bits(bits, selection):
#     sample = [] 
#     # Check if bits is a NumPy array
#     if isinstance(bits, np.ndarray):
#         # Use np.mod to make sure the bit we sample is always in the array range
#         selection = np.mod(selection, len(bits))
#         # Sample the bits using the corrected selection
#         sample = np.concatenate((sample, bits[selection]))
#         # Remove the sampled bits from the array
#         bits = np.delete(bits, selection, axis=0)
#     else:
#         # Use np.mod to make sure the bit we sample is always in the list range
#         selection = [np.mod(i, len(bits)) for i in selection]
#          # Pop the selected bits from the list
#         sample = [bits.pop(i) for i in selection]
#     return sample
def sample_bits(bits, selection):
    sample = []
    
    # Check if bits is a NumPy array
    if isinstance(bits, np.ndarray):
        # Use np.mod to make sure the bit we sample is always in the array range
        selection = np.mod(selection, len(bits))
        
        # Sample the bits using the corrected selection
        sample = np.concatenate((sample, bits[selection]))
        
        # Remove the sampled bits from the array
        bits = np.delete(bits, selection, axis=0)
    else:
        # Use np.mod to make sure the bit we sample is always in the list range
        selection = [np.mod(i, len(bits)) for i in selection]
        
        # Check if the indices in selection are valid
        valid_indices = [i for i in selection if 0 <= i < len(bits)]
        
        # Extract the selected bits from the list only if indices are valid
        sample = [bits[i] for i in valid_indices]
    
    return sample
```

```python
#calculate qber by measuring hamming_weight/total no. of bits
def calc_qber(bob_sample, alice_sample):
    hamming_weight = 0
    for i in range(len(bob_sample)):
        if bob_sample[i] != alice_sample[i]:
            hamming_weight = hamming_weight + 1
    qber = hamming_weight / len(bob_sample) *100
    return qber
```

```python
#calculate Shannon entropy
def calc_entropy(key):
    no_1s = 0
    no_0s = 0
    n=len(key)
    for i in range(len(key)):
        if key[i]== 0:
            no_0s += 1
        else:
            no_1s += 1
    p_no_1s=(no_1s)/len(key)
    p_no_0s=(no_0s)/len(key)
    
    # Ensure that the logarithm arguments are not zero
    p_no_1s = p_no_1s if p_no_1s != 0 else 1
    p_no_0s = p_no_0s if p_no_0s != 0 else 1
    
    h_key =  p_no_1s*math.log2(p_no_1s)+ p_no_0s*math.log2(p_no_0s)
    return -h_key 
```

```python
#calculate joint entropy
def joint_entropy(alice_key, bob_key):
    joint_key = list(zip(alice_key, bob_key)) #creates n-tuple
    total_entries = len(joint_key)
    
    # Count occurrences of each tuple
    n_00 = joint_key.count((0, 0))
    n_10 = joint_key.count((1, 0))
    n_01 = joint_key.count((0, 1))
    n_11 = joint_key.count((1, 1))
    
    #calculate probabilities
    p_00 = n_00/total_entries
    p_01 = n_01/total_entries
    p_10 = n_10/total_entries
    p_11 = n_11/total_entries
    
    # Ensure that the logarithm arguments are not zero
    p_00 = p_00 if p_00 != 0 else 1
    p_01 = p_01 if p_01 != 0 else 1
    p_10 = p_10 if p_10 != 0 else 1
    p_11 = p_11 if p_11 != 0 else 1
    
    h_joint = p_00 * math.log2(p_00) + p_10 * math.log2(p_10) + p_01 * math.log2(p_01) + p_11 * math.log2(p_11)
    return -h_joint
```

```python
def calc_mutual_info(alice_key, eve_key):
    h_mutual = calc_entropy(alice_key) + calc_entropy(eve_key) - joint_entropy(alice_key, eve_key)
    return h_mutual
```

```python
#Function to generate random bit-flip error for m bits in a list
def random_bit_flip(eve_key, m):
       # Ensure m is not greater than the length of the list
    m = min(m, len(eve_key))
    
    # Randomly select m indices to flip
    flip_indices = random.sample(range(len(eve_key)), m)

    # Flip the selected bits
    for index in flip_indices:
        eve_key[index] = 1 - eve_key[index]

    return eve_key
```

```python
#Simulation to check mutual information as well as qber
# Function to run the simulation
def run_simulation(iterations, min_sample_size, max_sample_size, min_bit_flip, max_bit_flip):
    results = []

    for iteration in tqdm_notebook(range(iterations), desc="Running simulations", leave=False):
        clear_output(wait=True)  # Clear the previous output
        print(f"Running simulation {iteration + 1}/{iterations}")

        for n in [1, 10, 100, 1000, 10000]:
            alice_key = np.random.randint(2, size=n)
            eve_key = []
            bob_key = []
            encoding_alice = []
            encoding_eve = []

            dps_qkd(alice_key, eve_key, encoding_alice)

            new_eve_key = random_bit_flip(eve_key, np.random.randint(min_bit_flip, max_bit_flip + 1))

            dps_qkd(new_eve_key, bob_key, encoding_eve)

            for sample_size in range(min_sample_size, max_sample_size + 1):
                bit_selection = np.random.randint(n, size=sample_size)
                bob_sample = sample_bits(bob_key, bit_selection)
                alice_sample = sample_bits(alice_key, bit_selection)

                qber = calc_qber(bob_sample, alice_sample)
                mutual_info_a_e = calc_mutual_info(alice_key, eve_key)
                mutual_info_a_b = calc_mutual_info(alice_key, bob_key)

                results.append({
                    'Iteration': iteration,
                    'Key_Length': n,
                    'Sample_Size': sample_size,
                    'Bit_Flip': len(new_eve_key) - len(bob_key),
                    'QBER': qber,
                    'Mutual_Info_A_E': mutual_info_a_e,
                    'Mutual_Info_A_B': mutual_info_a_b
                })

    return pd.DataFrame(results)

# Run the simulation with 1000 iterations, sample_size from 1 to 50, and random bit flip from 1 to 50
iterations = 1000
min_sample_size, max_sample_size = 1, 50
min_bit_flip, max_bit_flip = 1, 50

simulation_results = run_simulation(iterations, min_sample_size, max_sample_size, min_bit_flip, max_bit_flip)

# Save results to Excel file
simulation_results.to_excel('simulation_results.xlsx', index=False)

# Plot 3D graph
fig = plt.figure(figsize=(15, 8))
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122)

ax1.scatter(simulation_results['Sample_Size'], simulation_results['Bit_Flip'], simulation_results['QBER'], c='r', marker='o')
ax1.set_xlabel('Sample Size')
ax1.set_ylabel('Bit Flip')
ax1.set_zlabel('QBER')
ax1.set_title('3D Plot of QBER vs Sample Size and Bit Flip')

# Plot mutual information versus key length
ax2.plot(simulation_results['Key_Length'], simulation_results['Mutual_Info_A_E'], label='Mutual Info Alice-Eve', color='blue')
ax2.plot(simulation_results['Key_Length'], simulation_results['Mutual_Info_A_B'], label='Mutual Info Alice-Bob', color='red')
ax2.set_xlabel('Key Length (n)')
ax2.set_ylabel('Mutual Information')
ax2.set_title('Mutual Information vs Key Length')
ax2.legend()

plt.tight_layout()
plt.show()
```
